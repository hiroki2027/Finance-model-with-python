# データ収集前チェックリスト

1. ユースケースの整理  
   - 目的：シグナルを出す頻度と必要なカラムを明確化する  
   - 実装例：  
     - シグナル頻度（週次／日次／分足）を文書化  
     - 必要カラム（Open、High、Low、Close、Adj Close、Volume）を一覧化

2. ファイル命名・階層ルールの策定  
   - 目的：大量データを体系的に管理し混乱を防ぐ  
   - 実装例：  
     - `data/raw/{frequency}/{YYYY-MM-DD}/{TICKER}.parquet` の形式を標準化  


#### ここからは自動化(24時間・大量銘柄)のためなので今は必要なし

-----------------------------------------
3. バックアップおよび差分取得戦略の構築  
   - 目的：API障害やデータ欠損に備え、再取得を容易にする  
   - 実装例：  
     - ファイルハッシュを算出し変更検知  
     - `download_log.csv` に取得日時と行数を記録

4. 調整後価格の取り扱い  
   - 目的：株式分割や配当による価格飛びを排除する  
   - 実装例：  
     - `yf.download(..., auto_adjust=True)` を利用  
     - 後段処理用に「調整後価格フラグ」を用意

5. APIレート制限とリトライ設定  
   - 目的：API利用規約遵守と安定的なデータ取得を両立する  
   - 実装例：  
     - 連続200リクエストごとに2秒間のスリープを挿入  
     - 再試行（retry）および指数的バックオフを導入
-----------------------------------------
